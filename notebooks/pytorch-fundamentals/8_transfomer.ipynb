### Codebooks 
A codebook in quantization is a set of discrete embedding vectors used to replace continuous representations with the nearest prototype vector.
- continuous data → discrete
- compress data 
- learn discrete representations

Ex:   
codebook = [  
    e1 = [0.2, -1.0, 0.7]  
    e2 = [-0.5, 0.4, 1.1]  
    e3 = [0.3, -1.1, 0.9]  
]  
z = [0.3, -1.2, 0.8]

closest = e3

=> replace z by e3
