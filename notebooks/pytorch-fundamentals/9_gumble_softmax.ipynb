### Gumble Softmax: 
Gumbel-Softmax enables us to sample from a distribution while making it differentiable

Normal sample:  
- prob = [0.6,0.3,0.1] => result = [1,0,0]  
Gumbel-Softmax  
- prob = [0.6,0.3,0.1] => add noise => softmax => result ≈ [0.92,0.06,0.02]

A: 2.0 + noise (g1)
B: 1.0 + noise  (g2)
C: 0.5 + noise  (g3)

$$
softmax((logits + gumbel\_noise)/τ) \\

$$
u ~ Uniform(0,1)   
g = gumbel_noise = -log(-log(u))  
τ: temperature (hyperparameters, 0 < τ < ∞, usually 0.1->5) 

vd: sample   
u1 = 0.7 => g1 = -log(-log(0.7)) ≈ 1.03  
u2 = 0.2 => g2 = -log(-log(0.2)) ≈ -0.48  
u3 = 0.9 => g3 = -log(-log(0.9)) ≈ 2.25  
=> new values: 
logits = [2.0, 1.0, 0.5]  
g = [1.03, -0.48, 2.25 ]  
=> new_logits = logits + g = [3.03, 0.52, 2.3]
=> / T: temperature (maybe 0.5 → 1) => then softmax
